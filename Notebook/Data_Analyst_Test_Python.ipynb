```python
#Imports libraries

import numpy as np # for linear algebra functions
import pandas as pd # to process data in CSV file
import datetime as dt
```

```python
# importing csv files

artists = pd.read_csv('../data/artists.csv')

artworks = pd.read_csv('../data/artworks.csv')
```

```python
#Displaying top 5 rows of "artists" table

artists.head()
```

```python
# Making First letter of all values of Gender as capital for case sensitivity

artists['Gender'] = artists['Gender'].str.capitalize()

# Gender cannot be null until and unless it is a company, hence replacing null values with "Not Specified"
artists['Gender'].fillna("Not Specified", inplace = True) 
```

```python
#Displaying top 5 rows of "artworks" table

artworks.head()
```

```python
# dropping rows where Artist ID is Null

artworks = artworks.dropna(subset=['Artist ID'])

df = artworks.copy()

# in the dataset we have rows where in 'Artist ID' cell have multiple id's and names (collaboration artworks)
# I am trying to split those ID's and names and save then in new seperate rows , hence drooping unnecessary columns for this


df = df.drop(['Artist ID','Title' ,'Date', 'Medium', 'Dimensions', 'Credit', 'Catalogue', 'Department',
              'Object Number', 'Weight (kg)', 'Duration (s)'], axis=1)
```

```python
# creating two new empty columns

df['Artist ID'] =""
df['If_collab'] =""
df.head()
```

```python
#from artists table we will take artist id and name to form a dictionary. 
#we will use this dictionary to match the id's in df dataframe and add new rows to split collaboration artists in different rows

df_dict = pd.DataFrame(artists, columns = ['Name', 'Artist ID']) 
artistkey= list(df_dict['Artist ID'].unique())
dict_artist= dict()
for i in range(len(artistkey)):
    dict_artist[artistkey[i]]=df_dict[df_dict['Artist ID']==artistkey[i]]['Name'].values[0]
    
len(dict_artist)
```

```python
dict_artist
```

```python
# Here we are checking if the artist ID has "," in it. If yes, it is a collaborative art otherwise it is individual art.
# In case of individual art, we will assign the key of dict_artist ( which is artist ID) to new blank artist ID column and make
# if_collab comumn as "No". In case of collaborative art, index or row will be saved in "doubles list". 
# If there is any artist ID not present in dict_artist, row index will be appended in check_list .
# Here check_list is empty as all artist ID's are present in dictionary dict_artist



doubles= list()
check_list=list()
for i in range(len(artworks)):
    val=artworks.iloc[[i],[2]].values
    temp=val[0][0]
    if type(temp)!= int:
        if temp.find(',')== -1:
            x=int(temp)
            if x in dict_artist.keys():
                df.iloc[[i],[10]]=x
                df.iloc[[i],[11]]='No'
                
            else:
                df.iloc[[i],[10]]= np.nan
                check_list.append(i)

        else:
            doubles.append(i)
            df.iloc[[i],[10]]= np.nan
    else:
        if temp in dict_artist.keys():
            df.iloc[[i],[10]]= temp
            df.iloc[[i],[11]]='No'
        else:
            df.iloc[[i],[10]]= np.nan
            check_list.append(i)
            
```

```python
# dropping rows with Artist ID as null
df= df.dropna(subset=['Artist ID'])
```

```python
#Here we are using doubles list which has row indexes of collaboration artwork. Here we are splitting the artist ID 
# and creating new rows keeping all the values same except artist name and artist ID.
# Later we will append these rows in the dataframe. 

for i in doubles:
    val_id=artworks.iloc[[i],[2]].values[0][0]
    val_name=artworks.iloc[[i],[3]].values[0][0]
    list1=val_id.split(',')
    for j in range(len(list1)):
        try:
            temp_id= int(list1[j])
            if temp_id  in  dict_artist.keys():

                temp_name= dict_artist[temp_id]


                new_row = {'Artwork ID':artworks.iloc[[i],[0]].values[0][0],
                           'Name':temp_name,'Acquisition Date':artworks.iloc[[i],[7]].values[0][0],
                           'Classification':artworks.iloc[[i],[11]].values[0][0],
                           'Diameter (cm)':artworks.iloc[[i],[13]].values[0][0],
                           'Circumference (cm)':artworks.iloc[[i],[14]].values[0][0],
                           'Height (cm)':artworks.iloc[[i],[15]].values[0][0],
                           'Length (cm)':artworks.iloc[[i],[16]].values[0][0],
                           'Width (cm)':artworks.iloc[[i],[17]].values[0][0],
                           'Depth (cm)':artworks.iloc[[i],[18]].values[0][0],
                           'Artist ID': temp_id, 'If_collab': artworks.iloc[[i],[0]].values[0][0] }
                
                df = df.append(new_row, ignore_index=True)
            
            else:
                check_list.append(i)
        except Exception as e:
            print("Exception Occured at row number {}".format(i))
        
            print(e)
```

<font color = blue> **Question 3:** <br>
    **Which artist is created the most artwork by total surface area?** </font>

```python
# to calculate the surface area, we definitely need shape. As shape is not mentioned, we are assuming shapes and finding surface area.
# We are assuming only 2D and 3D shapes, however shapes can he irregular. For irregular shapes, we cannot find the surface area, hence 
# ignoring those rows.

df_sa = df_merged.copy()
df_sa['Shape'] =""
df_sa['Surface Area'] =""
```

```python
# dropping rows where diameter and circumference are not null and radius is different

from math import *
x=df_sa[pd.notnull(df_sa['Diameter (cm)']) & pd.notnull(df_sa['Circumference (cm)']) 
                & pd.isnull(df_sa['Length (cm)']) & pd.isnull(df_sa['Width (cm)']) 
                & pd.isnull(df_sa['Height (cm)'])& pd.isnull(df_sa['Depth (cm)'])]
indx = list(x.index.values)
df_sa = df_sa.drop(labels=indx , axis=0 )
```

```python
# dropping rows as only shape with height and diameter is cylinder or depth and diameter is cylinder. There is no shape having all three different values

x1=df_sa[pd.notnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Circumference (cm)']) 
                & pd.isnull(df_sa['Length (cm)']) & pd.isnull(df_sa['Width (cm)']) 
                & pd.notnull(df_sa['Height (cm)'])& pd.notnull(df_sa['Depth (cm)'])]
indx1 = list(x1.index.values)
df_sa = df_sa.drop(labels=indx1 , axis=0 )
```

```python
# shape with length and width and assuming it as a square/rectangle
x_len_wid=df_sa[pd.notnull(df_sa['Length (cm)']) & pd.notnull(df_sa['Width (cm)']) 
                & pd.isnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Circumference (cm)']) 
                & pd.isnull(df_sa['Height (cm)'])& pd.isnull(df_sa['Depth (cm)'])]

x_len_wid['Surface Area']= x_len_wid['Length (cm)']*x_len_wid['Width (cm)'] 
x_len_wid['Shape'] = '2 D'
```

```python
# shape with length and height and assuming it as a square/rectangle

x_len_height=df_sa[pd.notnull(df_sa['Length (cm)']) & pd.notnull(df_sa['Height (cm)']) 
                & pd.isnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Circumference (cm)']) 
                & pd.isnull(df_sa['Width (cm)'])& pd.isnull(df_sa['Depth (cm)'])]

x_len_height['Surface Area']= x_len_height['Length (cm)']*x_len_height['Height (cm)'] 
x_len_height['Shape'] = '2 D'
```

```python
# shape with width and height and assuming it as a square/rectangle

x_wid_height=df_sa[pd.notnull(df_sa['Width (cm)']) & pd.notnull(df_sa['Height (cm)']) 
                & pd.isnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Circumference (cm)']) 
                & pd.isnull(df_sa['Length (cm)'])& pd.isnull(df_sa['Depth (cm)'])]

x_wid_height['Surface Area']= x_wid_height['Width (cm)']*x_wid_height['Height (cm)'] 
x_wid_height['Shape'] = '2 D'
```

```python
# shape with circumference and assuming it as a circle

from math import *
x_circum=df_sa[pd.notnull(df_sa['Circumference (cm)']) & pd.isnull(df_sa['Width (cm)']) 
        & pd.isnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Length (cm)'])
        & pd.isnull(df_sa['Height (cm)'])& pd.isnull(df_sa['Depth (cm)'])]

x_circum['Surface Area']= pi * (x_circum['Circumference (cm)']/(2*pi))**2
x_circum['Shape'] = '2 D'
```

```python
# shape with circumference and assuming it as a shpere

x_sphere=df_sa[pd.notnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Width (cm)']) 
        & pd.isnull(df_sa['Circumference (cm)']) & pd.isnull(df_sa['Length (cm)'])
        & pd.isnull(df_sa['Height (cm)'])& pd.isnull(df_sa['Depth (cm)'])]

x_sphere['Surface Area']= 4 *pi * ((x_sphere['Diameter (cm)']/2)**2)
x_sphere['Shape'] = '3 D'
```

```python
# shape with diameter and height and assuming it as a cylinder

x_dia_height=df_sa[pd.notnull(df_sa['Diameter (cm)']) & pd.notnull(df_sa['Height (cm)']) 
        & pd.isnull(df_sa['Circumference (cm)']) & pd.isnull(df_sa['Length (cm)'])
        & pd.isnull(df_sa['Width (cm)'])& pd.isnull(df_sa['Depth (cm)'])]

x_dia_height['Surface Area']= (2 *pi *(x_dia_height['Diameter (cm)']/2)* x_dia_height['Height (cm)'])* (2*pi*((x_dia_height['Diameter (cm)']/2)**2))
x_dia_height['Shape'] = '3 D'
```

```python
# shape with diameter and depth and assuming it as a cylinder

x_dia_depth=df_sa[pd.notnull(df_sa['Diameter (cm)']) & pd.notnull(df_sa['Depth (cm)']) 
        & pd.isnull(df_sa['Circumference (cm)']) & pd.isnull(df_sa['Length (cm)'])
        & pd.isnull(df_sa['Width (cm)'])& pd.isnull(df_sa['Height (cm)'])]

x_dia_depth['Surface Area']= (2 *pi *(x_dia_depth['Diameter (cm)']/2)* x_dia_depth['Depth (cm)'])* (2*pi*((x_dia_depth['Diameter (cm)']/2)**2))
x_dia_depth['Shape'] = '3 D'
```

```python
# shape with circumference and height and assuming it as a cylinder

x_circum_height=df_sa[pd.notnull(df_sa['Circumference (cm)']) & pd.notnull(df_sa['Height (cm)']) 
        & pd.isnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Length (cm)'])
        & pd.isnull(df_sa['Width (cm)'])& pd.isnull(df_sa['Depth (cm)'])]

x_circum_height['Surface Area']= (2 *pi *(x_circum_height['Circumference (cm)']/(2*pi))* x_circum_height['Height (cm)'])* (2*pi*((x_circum_height['Circumference (cm)']/(2*pi))**2))
x_circum_height['Shape'] = '3 D'
```

```python
# shape with circumference and depth and assuming it as a cylinder

x_circum_depth=df_sa[pd.notnull(df_sa['Circumference (cm)']) & pd.notnull(df_sa['Depth (cm)']) 
        & pd.isnull(df_sa['Diameter (cm)']) & pd.isnull(df_sa['Length (cm)'])
        & pd.isnull(df_sa['Width (cm)'])& pd.isnull(df_sa['Height (cm)'])]

x_circum_depth['Surface Area']= (2 *pi *(x_circum_depth['Circumference (cm)']/(2*pi))* x_circum_depth['Depth (cm)'])* (2*pi*((x_circum_depth['Circumference (cm)']/(2*pi))**2))
x_circum_depth['Shape'] = '3 D'
```

```python
# shape with length, width and height and assuming it as a cuboid

x_len_width_height =df_sa[pd.notnull(df_sa['Length (cm)']) & pd.notnull(df_sa['Width (cm)']) 
        & pd.notnull(df_sa['Height (cm)']) & pd.isnull(df_sa['Depth (cm)'])
        & pd.isnull(df_sa['Circumference (cm)'])& pd.isnull(df_sa['Diameter (cm)'])]

x_len_width_height['Surface Area']= (2*x_len_width_height['Length (cm)']*x_len_width_height['Width (cm)'])+ (2*x_len_width_height['Length (cm)']*x_len_width_height['Height (cm)']) + (2*x_len_width_height['Height (cm)']*x_len_width_height['Width (cm)'])
x_len_width_height['Shape'] = '3 D'
```

```python
df_sa_final = pd.concat([x_len_wid, x_len_height, x_wid_height , x_circum , x_sphere , x_dia_height , x_dia_depth , x_circum_height , x_circum_depth ,x_len_width_height ], ignore_index=True)
df_sa_final
```

```python
# to get ID of the artist whose sum of surface area is max

ID = df_sa_final.groupby('Artist ID')
total_sa = ID.apply(lambda x: x['Surface Area'].sum())
total_sa.sort_values(ascending=False, inplace=True)
total_sa
```
